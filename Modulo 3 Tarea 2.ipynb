### RegPlot : Now we going to see if there is any correlation between the size of the table (people), and the tips.

```python
#Setting the size of the chart
plt.figure(figsize =(20, 8))

#Setting the type of chart
sns.regplot(data=df, x='size', y='tip', fit_reg=True, line_kws=dict(color="red"))

#Setting others parameters
plt.title("Regplot",fontsize=24)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel("Size", fontsize=18)
plt.ylabel("Tips", fontsize=18)

plt.grid(color='g', linestyle='-', linewidth=0.5)

#plt.savefig('Scatterplot.png')

plt.show()
```

```python
#Setting the size of the chart
plt.figure(figsize =(20, 8))

#Setting the type of chart
sns.regplot(data=df, x='total_bill', y='tip', fit_reg=True, line_kws=dict(color="red"))

#Setting other parameters
plt.title("Regplot",fontsize=24)
plt.xlabel("Total Bill", fontsize=18)
plt.ylabel("Tips", fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.grid(color='g', linestyle='-', linewidth=0.5)

#plt.savefig('Scatterplot.png')
plt.show()
```

### In both charts we can see a positive and strong correlation between the size of the table and the tips (first chart), that means when the size of the table is higher the tips increase. The same happen in the second chart. We can see a positive correlation and strong correlation between the total bill and the tips, when the bill is higher the tips also is higher.

Remember that the correlation is the relationship between the variables, a number between -1 and 1. when the number its positive means that when one is up the others also increment, and when its negative its the opposite, when one variable is up the others are down. 1 means a perfect positive correlation.

### More than 3 variables. In the Chart Below

```python
#Variable
data = df

#Setting the size of the chart.
plt.figure(figsize = (15,10))

#Setting the type of chart.
ax = sns.scatterplot(data=data, x="total_bill", y="tip",\
                    size="size", sizes=(300, 1500),\
                    alpha = 0.5,
                    hue="smoker",
                    hue_order=['Yes', 'No'],
                    palette="rainbow",
                    edgecolor="black",
                    )

#Setting others parameters
plt.legend(title="Size", loc='best', fontsize= 16)    
plt.legend(bbox_to_anchor=(1, 1),loc='best', fontsize=11, borderaxespad=2)

plt.title("Bubble Chart:", fontsize=20)
plt.xlabel("Total Bill", fontsize= 18, color="black")
plt.ylabel("Tip", fontsize= 20, color="black")
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)


ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.yaxis.grid(True)
ax.xaxis.grid(True)
#plt.savefig('bubble_chart.png')


plt.show()
```

### In this chart we notice the relationship between the total bill, the tip, if is smoker or not, and the size of the table. The total bill is in the Y axe. The tip is represented in the Y axe. The Smoker variable is representend by the color of the bubble meanwhile the size of the table (number of people), is represented by the "size" of the bubble.
As we can see higher is the bill and the size, higher is the tip, and we notice the distribution between smokers and number of people.
